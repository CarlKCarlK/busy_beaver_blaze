# Interactive Turing Machine Visualization

This notebook demonstrates the interactive visualization of Turing Machines.

````python
from busy_beaver_blaze import Visualizer, BB5_CHAMP, BB6_CONTENDER  # noqa: F401
````

## Example: BB5 Champion with Early Stop

Run the BB5 champion machine up to 10,000,000 steps with live updates.

````python
# Create visualizer (defaults to 800x600 and colorful palette)
visualizer = Visualizer(
    program=BB5_CHAMP
)

# Visualize with early stop
visualizer.run_live(
    early_stop=10_000_000,
    caption="BB5 Champion"
)
````

## Example: Indefinite Running (Stop with Ctrl+C)

Run without an early_stop limit. Use the stop button or Ctrl+C to interrupt.

````python
# Create fresh visualizer (uses defaults)
visualizer = Visualizer(
    program=BB5_CHAMP
)

# Run indefinitely - click 'Interrupt' button or press Ctrl+C to halt
visualizer.run_live(
    caption="BB5 - Click 'Interrupt' to Halt"
)
````

## Example: Custom Colors

Create a machine with custom color palette.

````python
# Create visualizer with custom colors
visualizer = Visualizer(
    program=BB5_CHAMP,
    colors=["#FFFFFF", "#FF5500", "#FFFF00", "#FF00FF", "#00FFFF"]  # White, orange, yellow, magenta, cyan
)

visualizer.run_live(
    early_stop=50_000,
    caption="BB5 with Custom Colors"
)
````

## Example: User-defined Program (BB6 Contender)

You can create your own Turing machine programs using any of the common string formats. You can also save to a file.

For example, BB6 Contender program with the image saved to a file.

```text
	A	B	C	D	E	F
0	1RB	1RC	1LC	0LE	1LF	0RC
1	0LD	0RF	1LA	1RH	0RB	0RE
```

You can pass either the built-in `BB6_CONTENDER` constant or the string literal above.

````python
from pathlib import Path
import tempfile
from busy_beaver_blaze import Visualizer

out_file = Path(tempfile.gettempdir()) / "busy_beaver_blaze_images" / "bb6_200K.png"

# Create and step the visualizer to a point of interest
visualizer = Visualizer(
    program="""
	A	B	C	D	E	F
0	1RB	1RC	1LC	0LE	1LF	0RC
1	0LD	0RF	1LA	1RH	0RB	0RE
"""
)
# Render the frame and optionally save it to disk
img, step_count = visualizer.run(early_stop=200_000, out_file=out_file)

# Display the image in the notebook
display(img)
print(f"Captured step count: {step_count:,}")
print(f"Saved to: {out_file}")
````
